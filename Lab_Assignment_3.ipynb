```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# Task
Data: [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]<br>
Hypothesized mean: 5.5 <br>

Tasks:<br>
- Calculate the t-statistic and p-value.<br>
- Interpret and visualize the result at a 0.05 significance level.<br>


```python
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
hypothesized_mean = 5.5
significance_level = 0.05

t_statistic, p_value = stats.ttest_1samp(data, hypothesized_mean)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

if p_value < significance_level:
    print("Reject the null hypothesis. The sample mean is significantly different from the hypothesized mean.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the sample mean is significantly different from the hypothesized mean.")

plt.figure()
plt.bar(1, np.mean(data), color="lime", label='Sample Mean')
plt.errorbar(1, np.mean(data), yerr=np.std(data), fmt='k', linewidth=1.5)
plt.axhline(y=hypothesized_mean, color='r', linestyle='--', linewidth=1.5, label='Hypothesized Mean')
plt.title('Sample Mean vs Hypothesized Mean')
plt.xlim(0.5,1.5)
plt.xlabel('Sample')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
