```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Task-1:** Generate two vectors with 15 random floats each, plot them on the same graph, label the axes, and add a title.

```python
v1 = np.random.randint(4, 37, size=(4, 4))
v2 = np.random.randint(1, 29, size=(4, 4))

v1_flat = v1.flatten()
v2_flat = v2.flatten()

x = np.arange(len(v1_flat))

plt.plot(x, v1_flat, marker='o', linestyle='-', label="Vector A")
plt.plot(x, v2_flat, marker='s', linestyle='--', label="Vector B")

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Two Random Vectors')

plt.legend()

plt.show()
```

### **Task-2:** Create a 4x4 matrix with random values, visualize it as a heatmap, and label the rows and columns.

```python
matrix = np.random.randint(4, 28, size=(4, 4))

row_labels = ['Row 1', 'Row 2', 'Row 3', 'Row 4']
col_labels = ['Col 1', 'Col 2', 'Col 3', 'Col 4']

sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.xticks(np.arange(len(col_labels)), col_labels)
plt.yticks(np.arange(len(row_labels)), row_labels)

plt.title('Heatmap of 4x4 Matrix')
plt.show()
```

### **Task-3:** Generate two 4x4 matrices, perform addition, subtraction, and multiplication, visualize the results using bar plots, and discuss the effects.

```python
A = np.random.randint(1, 10, (4, 4))
B = np.random.randint(1, 10, (4, 4))

add = A + B
sub = A - B
mult = np.dot(A, B)

print('Summation \n',add)
print('Subtraction \n',sub)
print('Multiplication \n',mult)

add_flat = add.flatten()
sub_flat = sub.flatten()
mult_flat = mult.flatten()

fig, axs = plt.subplots(1, 3, figsize=(20, 5))

axs[0].bar(range(len(add_flat)), add_flat, color='orange')
axs[0].set_title("Matrix A + B")
axs[0].set_ylim([0, 40])

axs[1].bar(range(len(sub_flat)), sub_flat, color='red')
axs[1].set_title("Matrix A - B")
axs[1].set_ylim([-20, 20])

axs[2].bar(range(len(mult_flat)), mult_flat, color='purple')
axs[2].set_title("Matrix A * B")
axs[2].set_ylim([0, 200])

plt.tight_layout()
plt.show()
```

#### **Discussion**
**Matrix Addition:** In the plot of the addition, we can see that each element is simply the sum of the corresponding elements of the two matrices. Since the numbers range from 1 to 9, the result will have values between 2 and 18.

**Matrix Subtraction:** The subtraction plot shows the difference between corresponding elements. The results can be both positive and negative depending on the relative values of A and B.

**Matrix Multiplication:** Matrix multiplication results in a more complex pattern compared to addition and subtraction. The values can vary more dramatically since each element is a sum of products of corresponding elements from rows of A and columns of B. This could lead to much larger values, especially if the original matrices contain large numbers.
